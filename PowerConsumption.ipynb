```
import seaborn as sns
```

```
plt.figure(figsize=(15,5))
plt.title('GHG Emissions Intensity(kgCO2e/ft2) for all Building Types')
sns.barplot(x='BuildingType',y='GHGEmissionsIntensity(kgCO2e/ft2)',data=df_2015,palette='rainbow')
```

Thus the maximum intensity is from Campus Building.

```
plt.figure(figsize=(22,5))
plt.title('GHG Emissions Intensity(kgCO2e/ft2) for all Neighbourhoods for the year 2015')
sns.barplot(x='Neighborhood',y='GHGEmissionsIntensity(kgCO2e/ft2)',data=df_2015,palette='rainbow')
```

```
plt.figure(figsize=(25,5))
plt.title('Energy Star Score for Every neighbourhood for the year 2015')
sns.lineplot(x='Neighborhood',y='ENERGYSTARScore',data=df_2015)
plt.figure(figsize=(25,5))
plt.title('Energy Star Score for Every neighbourhood for the year 2016')
sns.lineplot(x='Neighborhood',y='ENERGYSTARScore',data=df_2016)
```

```
plt.figure(figsize=(15,15))
plt.title('Heatmap for the year 2015')
sns.heatmap(data=df_2015.corr(),annot=True, cmap='YlGnBu')
```
